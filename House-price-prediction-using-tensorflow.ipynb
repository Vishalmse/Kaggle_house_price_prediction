```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Dropping feature which have more missing value then 50% of dataset

```python
data.drop(['Id','Alley','PoolQC','Fence','MiscFeature'], axis=1, inplace=True)
```

## Dealing with missing values

```python
data['LotFrontage'] = data['LotFrontage'].fillna(data['LotFrontage'].mean())
data['MasVnrType'] = data['MasVnrType'].fillna(data['MasVnrType'].mode()[0])
data['MasVnrArea'] = data['MasVnrArea'].fillna(data['MasVnrArea'].mean())
data['BsmtQual'] = data['BsmtQual'].fillna(data['BsmtQual'].mode()[0])
data['BsmtCond'] = data['BsmtCond'].fillna(data['BsmtCond'].mode()[0])
data['BsmtExposure'] = data['BsmtExposure'].fillna(data['BsmtExposure'].mode()[0])
data['BsmtFinType1'] = data['BsmtFinType1'].fillna(data['BsmtFinType1'].mode()[0])
data['BsmtFinType2'] = data['BsmtFinType2'].fillna(data['BsmtFinType2'].mode()[0])
data['Electrical'] = data['Electrical'].fillna(data['Electrical'].mode()[0])
data['FireplaceQu'] = data['FireplaceQu'].fillna(data['FireplaceQu'].mode()[0])
data['GarageType'] = data['GarageType'].fillna(data['GarageType'].mode()[0])
data['GarageYrBlt'] = data['GarageYrBlt'].fillna(data['GarageYrBlt'].mean())
data['GarageFinish'] = data['GarageFinish'].fillna(data['GarageFinish'].mode()[0])
data['GarageQual'] = data['GarageQual'].fillna(data['GarageQual'].mode()[0])
data['GarageCond'] = data['GarageCond'].fillna(data['GarageCond'].mode()[0])
```

```python
clean_train_data = data.copy()
clean_train_data.shape
```

## Merging train and test data to get dummies for both data

```python
clean_test_data = pd.read_csv('clean_test_data')
```

```python
final_df = pd.concat([clean_train_data,clean_test_data], axis=0)
final_df.shape
```

## creating dummy variable



```python
def get_dummy(list_columns, final_df):
    df_final = final_df
    i = 0
    for field in list_columns:
        print(field)
        df1 = pd.get_dummies(final_df[field], drop_first=True)
        
        final_df.drop([field], axis=1, inplace=True)
        if i == 0:
            df_final= df1.copy()
        else:
            final_df= pd.concat([final_df, df1], axis=1)
        i += 1
    
    final_df = pd.concat([final_df, df_final], axis=1)
    
    return final_df    
```

```python
# make list of all catogerical data
columns = []
for i in clean_train_data:
    if clean_train_data[i].dtypes == "O":
        columns.append(i)
len(columns)
```

```python
final_df = get_dummy(columns, final_df)
```

```python
final_df = final_df.loc[:, ~final_df.columns.duplicated()]
final_df.shape
```

```python
df_train = final_df.iloc[:1460,:]
df_test = final_df.iloc[1460:,:]
```

```python
df_test.drop(['SalePrice'], axis=1, inplace=True)
```

## Split train,validation,test data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
train_input = df_train.drop('SalePrice', axis=1)
train_output = df_train['SalePrice']
```

```python
scaler = StandardScaler()
scaled_train = scaler.fit(train_input).transform(train_input)
```

```python
X_train_val, X_test, y_train_val, y_test = train_test_split(scaled_train, train_output, test_size=0.1,
                                                            random_state=1215)

X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.2,
                                                            random_state=1215)
print('X_train shape = ',X_train.shape)
print('X_val shape = ', X_val.shape)
print('X_test shape = ', X_test.shape)
print('y_train shape = ', y_train.shape)
print('y_val shape = ', y_val.shape)
print('y_test shape = ', y_test.shape)
```
